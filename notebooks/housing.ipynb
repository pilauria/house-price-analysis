```python
#import and read the data
import pandas as pd
df = pd.read_csv('housing.csv')
df.head()
```

```python
#info about the data
df.info()
#describe the data
df.describe() 

```

```python
# Option 1: Use size for Garage Area and color for Overall Quality
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")

sns.relplot(data=df, 
           x='Gr Liv Area', 
           y='SalePrice',
           hue='Overall Qual',
           palette='rocket_r',
           size='Garage Area',    # Now size represents garage area
           sizes=(1, 300),       # Adjust size range as needed
           height=8,
           aspect=1.5)

# Option 2: Create a subplot grid to show both relationships
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# First plot: Living Area vs Price
sns.scatterplot(data=df, 
                x='Gr Liv Area', 
                y='SalePrice',
                hue='Overall Qual',
                palette='rocket_r',
                ax=ax1)
ax1.set_title('Price vs Living Area')

# Second plot: Garage Area vs Price
sns.scatterplot(data=df, 
                x='Garage Area', 
                y='SalePrice',
                hue='Overall Qual',
                palette='rocket_r',
                ax=ax2)
ax2.set_title('Price vs Garage Area')

plt.tight_layout()
```

```python
sns.set_theme(style="whitegrid", context="talk")

sns.relplot(data=df, 
           x='Gr Liv Area', 
           y='SalePrice',
           hue='Overall Qual',
           palette='rocket_r',
           size='Garage Area',    
           sizes=(20, 400),
           style='Rooms',         # Add different markers based on rooms
           markers={'6 rooms or less': 'o', '7 rooms or more': '^'},  # circle and triangle
           height=8,
           aspect=1.5)

# Add a title
plt.title('House Prices: Living Area, Quality, Garage Size, and Rooms')
```

```python
# Create a year category (assuming 2000 as the splitting point)
df['Year Category'] = df['Year Built'].apply(lambda x: '2000 or newer' if x >= 2000 else '1999 or older')

# Create two separate plots using FacetGrid
sns.set_theme(style="whitegrid", context="talk")

g = sns.relplot(data=df, 
           x='Gr Liv Area', 
           y='SalePrice',
           hue='Overall Qual',
           palette='rocket_r',
           size='Garage Area',    
           sizes=(20, 400),
           style='Rooms',         
           markers={'6 rooms or less': 'o', '7 rooms or more': '^'},
           col='Year',    # Split plots by Year Category
           height=8,
           aspect=1.2)

# Adjust titles
g.fig.suptitle('House Prices by Age: Living Area, Quality, Garage Size, and Rooms', y=1.05)
```
